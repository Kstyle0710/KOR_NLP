# WS 토크나이징

```python
import pandas as pd
from tqdm import tqdm
import nltk
from nltk.tokenize import RegexpTokenizer
from konlpy.tag import Mecab
mecab = Mecab(dicpath=r"C:/mecab/mecab-ko-dic")
nltk.download()   #'punkt'
```

```python
stop_word = "전 난 일 걸 뭐 줄 만 건 작업 분 위 개 끝 송 잼 이거 부 동 번 중 듯 차 때 게 내 말 나 수 거 점 것 등 측 의 급 후 간 단 시 곳"
stop_word = stop_word.split(' ')
print(stop_word)
```

```python
df = pd.read_excel('./data/작업표준 목록_조선외.xlsx')
target = df["표준서명"]
```

```python
targets = []
results = []
```

```python
for sentence in tqdm(target):
    result = []
    tokenizer = RegexpTokenizer("[\w]+")
    tokens = tokenizer.tokenize(sentence)
    for token in tokens:
        if token not in stop_word:
            result.append(token)
    targets.append(sentence)
    results.append(result)
```

```python
results[0]
```

```python
df.to_excel("./result/ko_nlp_1214(3).xlsx", encoding='utf-8-sig')
```

# 부서별 토큰 그룹핑

```python
df["tokenized"] = results
df.head()
```

```python
i = '기계가공부'
t1 = df.loc[df["기안자"] == i, 'tokenized']
t1.head()
```

```python
t2 = t1.values.tolist()
t2[0][1]
```

```python
final ={}
depts = df["기안자"].unique()
for i in depts:
    temp_list =[]
    t7 = df.loc[df["기안자"] == i, 'tokenized']
    t8 = t7.values.tolist()
    for j in t8:
        for m in j:
            if m not in temp_list:
                temp_list.append(m)
            else:
                pass

    final[i] = temp_list
print(final)
```

```python
key = final.keys()
len(key)
```

```python
value = final.values()
len(value)
```

```python
summary = [key, value]
summary
```

```python
df3 = pd.DataFrame(summary)
df4 = df3.T
df4
```

# N-gram

```python
from nltk import ngrams
```

```python
sentence = "외주 제작품 도장시공 표준 요령"
```

```python
unigram = list(ngrams(sentence.split(), 1))
unigram
```

```python
bigram = list(ngrams(sentence.split(), 2))
bigram
```

```python
trigram = list(ngrams(sentence.split(), 3))
trigram
```
